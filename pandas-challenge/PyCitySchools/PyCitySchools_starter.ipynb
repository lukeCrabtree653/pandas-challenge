```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
schoolList = school_data_complete["school_name"].unique()
schoolCount = len(schoolList)
studentList = school_data_complete["student_name"]
studentCount = len(studentList)
totalBudget = school_data_complete["budget"].unique().sum()
totalBudget = "${:,.2f}".format(totalBudget)
averageMS = school_data_complete["math_score"].mean()
averageRS = school_data_complete["reading_score"].mean()
passingMath = school_data_complete.loc[school_data_complete["math_score"] >= 70]
passingMathCount = passingMath["math_score"].count()
passingMathPercentage = 100 * passingMathCount / studentCount
passingReading = school_data_complete.loc[school_data_complete["reading_score"] >= 70]
passingReadingCount = passingReading["reading_score"].count()
passingReadingPercentage = 100 * passingReadingCount / studentCount
passingBoth = school_data_complete.loc[(school_data_complete["math_score"] >= 70) & (school_data_complete["reading_score"] >= 70)]
passingBothCount = passingBoth["math_score"].count()
passingBothPercentage = 100 * passingBothCount / studentCount
passingBothPercentage
DistrictSummary_df = pd.DataFrame()
DistrictSummary_df["Total Schools"] = [schoolCount]
DistrictSummary_df["Total Students"] = [studentCount]
DistrictSummary_df["Total Budget"] = [totalBudget]
DistrictSummary_df["Average Math Score"] = [averageMS]
DistrictSummary_df["Average Reading Score"] = [averageRS]
DistrictSummary_df["% Passing Math"] = [passingMathPercentage]
DistrictSummary_df["% Passing Reading"] = [passingReadingPercentage]
DistrictSummary_df["% Overall Passing"] = [passingBothPercentage]
DistrictSummary_df.head()
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
spendingScores = pd.DataFrame()
dataSchoolSpend = pd.DataFrame()
passingMathDataSchoolSpend = pd.DataFrame()
passingReadingDataSchoolSpend = pd.DataFrame()
passingBothDataSchoolSpend = pd.DataFrame()
dataSchoolSpend = psb
passingMathDataSchoolSpend = psb
passingReadingDataSchoolSpend = psb
passingBothDataSchoolSpend = psb

bins1 = [0, 584.99, 629.99, 644.99,679.99]
group_names1 = ["<$585", "$585-630", "$630-645","$645-680"]

dataSchoolSpend["Spending Ranges (Per Student)"] = pd.cut(dataSchoolSpend["Per Student Budget"],bins1,labels = group_names1,include_lowest = True)
passingMathDataSchoolSpend["Spending Ranges (Per Student)"] = pd.cut(passingMathDataSchoolSpend["Per Student Budget"],bins1,labels = group_names1,include_lowest = True)
passingReadingDataSchoolSpend["Spending Ranges (Per Student)"] = pd.cut(passingReadingDataSchoolSpend["Per Student Budget"],bins1,labels = group_names1,include_lowest = True)
passingBothDataSchoolSpend["Spending Ranges (Per Student)"] = pd.cut(passingBothDataSchoolSpend["Per Student Budget"],bins1,labels = group_names1,include_lowest = True)

dataSchoolSpend = dataSchoolSpend.groupby(["Spending Ranges (Per Student)"])
passingMathDataSchoolSpend = passingMathDataSchoolSpend.groupby(["Spending Ranges (Per Student)"])
passingReadingDataSchoolSpend = passingReadingDataSchoolSpend.groupby(["Spending Ranges (Per Student)"])
passingBothDataSchoolSpend = passingBothDataSchoolSpend.groupby(["Spending Ranges (Per Student)"])

spendingScores["Average Math Score"] = dataSchoolSpend["Average Math Score"].mean()
spendingScores["Average Reading Score"] = dataSchoolSpend["Average Reading Score"].mean()
spendingScores["% Passing Math"] = passingMathDataSchoolSpend["% Passing Math"].sum() / dataSchoolSpend["Total Students"].count()
spendingScores["% Passing Reading"] = passingReadingDataSchoolSpend["% Passing Reading"].sum() / dataSchoolSpend["Total Students"].count()
spendingScores["% Overall Passing"] = passingBothDataSchoolSpend["% Overall Passing"].sum() / dataSchoolSpend["Total Students"].count()

spendingScores
```
